```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from keras import layers, Model, optimizers, callbacks
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import BertTokenizer, TFBertModel
import tensorflow.compat.v1 as tf_compat

print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {keras.__version__}")

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load the dataset
# Assuming you have a CSV file with 'poem' and 'topic' columns
df = pd.read_csv('data.csv')

# Display dataset info
print(f"Dataset shape: {df.shape}")
print(f"Theme distribution:\n{df['topic'].value_counts()}")

# Preprocess the data
# Encode the labels
label_encoder = LabelEncoder()
df['encoded_topic'] = label_encoder.fit_transform(df['topic'])
num_classes = len(label_encoder.classes_)
print(f"Classes: {label_encoder.classes_}")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    df['poem'].values, 
    df['encoded_topic'].values,
    test_size=0.2,
    random_state=42,
    stratify=df['encoded_topic']
)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

# Initialize BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
max_length = 128  # Can be adjusted based on your poems' length

# Tokenize and prepare the data
def encode_poems(texts, max_length=128):
    input_ids = []
    attention_masks = []
    
    for text in texts:
        encoded = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='tf'
        )
        
        input_ids.append(encoded['input_ids'][0])
        attention_masks.append(encoded['attention_mask'][0])
    
    return np.array(input_ids), np.array(attention_masks)

# Encode training and testing data
train_input_ids, train_attention_masks = encode_poems(X_train, max_length)
test_input_ids, test_attention_masks = encode_poems(X_test, max_length)

# Convert outputs to categorical if needed
y_train_array = y_train.astype(np.int32)
y_test_array = y_test.astype(np.int32)

# For TF 2.18.0 and Keras 3.6.0, we need a different approach
# We'll use a functional API approach with TensorFlow tensors converted from Keras tensors

# First we'll create a custom wrapper layer for BERT
class BERTLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(BERTLayer, self).__init__(**kwargs)
        self.bert = TFBertModel.from_pretrained('bert-base-uncased', from_pt=False)
        
    def build(self, input_shape):
        self.built = True
    
    def call(self, inputs):
        # Convert Keras tensors to TF tensors
        input_ids_tf = tf.convert_to_tensor(inputs[0])
        attention_mask_tf = tf.convert_to_tensor(inputs[1])
        
        outputs = self.bert(input_ids_tf, attention_mask=attention_mask_tf)
        return outputs[0]  # Return the sequence output
    
    def get_config(self):
        config = super(BERTLayer, self).get_config()
        return config

# Build the model with TF 2.18.0 / Keras 3.6.0 compatibility
def build_model(num_classes):
    # Input layers
    input_ids = layers.Input(shape=(max_length,), dtype=tf.int32, name='input_ids')
    attention_mask = layers.Input(shape=(max_length,), dtype=tf.int32, name='attention_mask')
    
    # BERT layer
    bert_layer = BERTLayer()
    sequence_output = bert_layer([input_ids, attention_mask])
    
    # BiLSTM layer
    bilstm = layers.Bidirectional(layers.LSTM(256, return_sequences=False))(sequence_output)
    
    # Fully connected layers
    x = layers.Dense(256, activation='relu')(bilstm)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    
    # Output layer
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    # Create model
    model = Model(inputs=[input_ids, attention_mask], outputs=outputs)
    
    # Freeze BERT layers for initial training
    bert_layer.trainable = False
    
    return model, bert_layer

# Create and compile the model
model, bert_layer = build_model(num_classes)

# TF 2.18.0 compatible optimizer and loss
model.compile(
    optimizer=optimizers.Adam(learning_rate=3e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display model summary
model.summary()

# Setup callbacks
model_callbacks = [
    callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
    callbacks.ModelCheckpoint('bert_bilstm_poem_classifier.keras', monitor='val_accuracy', save_best_only=True, verbose=1)
]

# Initial training with frozen BERT layers
print("Phase 1: Training with frozen BERT layers...")
history_frozen = model.fit(
    [train_input_ids, train_attention_masks],
    y_train_array,
    validation_split=0.1,
    epochs=5,
    batch_size=8,  # Reduced batch size for memory efficiency
    callbacks=model_callbacks,
    verbose=1
)

# Unfreeze BERT layers for fine-tuning
print("Phase 2: Fine-tuning the entire model...")
bert_layer.trainable = True

# Recompile with a lower learning rate
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),  # Lower learning rate for fine-tuning
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Fine-tune the entire model
history_fine_tuned = model.fit(
    [train_input_ids, train_attention_masks],
    y_train_array,
    validation_split=0.1,
    epochs=5,
    batch_size=4,  # Even smaller batch size for fine-tuning
    callbacks=model_callbacks,
    verbose=1
)

# Plot training history
def plot_training_history(history_frozen, history_fine_tuned):
    # Combine histories
    epochs_frozen = len(history_frozen.history['accuracy'])
    epochs_fine_tuned = len(history_fine_tuned.history['accuracy'])
    
    acc = history_frozen.history['accuracy'] + history_fine_tuned.history['accuracy']
    val_acc = history_frozen.history['val_accuracy'] + history_fine_tuned.history['val_accuracy']
    loss = history_frozen.history['loss'] + history_fine_tuned.history['loss']
    val_loss = history_frozen.history['val_loss'] + history_fine_tuned.history['val_loss']
    epochs = range(1, epochs_frozen + epochs_fine_tuned + 1)
    
    # Plot the metrics
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'b-', label='Training Accuracy')
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    plt.axvline(x=epochs_frozen, color='green', linestyle='--', 
               label='Start Fine-tuning')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.axvline(x=epochs_frozen, color='green', linestyle='--', 
               label='Start Fine-tuning')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

# Plot the training history
plot_training_history(history_frozen, history_fine_tuned)

# Evaluate the model on test data
print("Evaluating model on test data...")
y_pred_probs = model.predict([test_input_ids, test_attention_masks])
y_pred = np.argmax(y_pred_probs, axis=1)

# Print classification report
print("\nClassification Report:")
print(classification_report(
    y_test_array, 
    y_pred, 
    target_names=label_encoder.classes_
))

# Plot confusion matrix
cm = confusion_matrix(y_test_array, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(
    cm, 
    annot=True, 
    fmt='d', 
    cmap='Blues',
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()

# Save the model weights 
# TF 2.18.0 & Keras 3.6.0 might have issues with full model saving with custom layers
model.save_weights('bert_bilstm_poem_classifier_weights.h5')
print("Model weights saved successfully.")

# Function to save the model configuration for later reconstruction
def save_model_config():
    model_config = model.get_config()
    with open('model_config.json', 'w') as f:
        import json
        json.dump(model_config, f)
    print("Model configuration saved successfully.")

# Try to save model config
try:
    save_model_config()
except Exception as e:
    print(f"Error saving model config: {e}")

# Function to predict theme for a new poem
def predict_theme(poem_text):
    # Tokenize the poem
    input_id, attention_mask = encode_poems([poem_text])
    
    # Make prediction
    prediction = model.predict([input_id, attention_mask])
    predicted_class = np.argmax(prediction, axis=1)[0]
    predicted_theme = label_encoder.classes_[predicted_class]
    confidence = prediction[0][predicted_class]
    
    return {
        'theme': predicted_theme,
        'confidence': float(confidence),
        'all_probabilities': {
            theme: float(prob) 
            for theme, prob in zip(label_encoder.classes_, prediction[0])
        }
    }

# Example usage
sample_poem = """
Roses are red,
Violets are blue,
Sugar is sweet,
And so are you.
"""

print("\nTesting prediction function:")
result = predict_theme(sample_poem)
print(f"Predicted theme: {result['theme']} with confidence: {result['confidence']:.2f}")
print("All theme probabilities:")
for theme, prob in sorted(result['all_probabilities'].items(), key=lambda x: x[1], reverse=True):
    print(f"  {theme}: {prob:.4f}")

# Create a simple interface for interactive predictions
class PoemClassifier:
    def __init__(self, model, label_encoder, tokenizer, max_length=128):
        self.model = model
        self.label_encoder = label_encoder
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def predict(self, poem_text):
        # Tokenize the poem
        input_id, attention_mask = encode_poems([poem_text], self.max_length)
        
        # Make prediction
        prediction = self.model.predict([input_id, attention_mask])
        predicted_class = np.argmax(prediction, axis=1)[0]
        predicted_theme = self.label_encoder.classes_[predicted_class]
        confidence = prediction[0][predicted_class]
        
        # Get all probabilities
        all_probs = {
            theme: float(prob) 
            for theme, prob in zip(self.label_encoder.classes_, prediction[0])
        }
        
        # Sort probabilities
        sorted_probs = sorted(all_probs.items(), key=lambda x: x[1], reverse=True)
        
        return {
            'theme': predicted_theme,
            'confidence': float(confidence),
            'all_probabilities': all_probs,
            'sorted_probabilities': sorted_probs
        }
    
    def interactive_mode(self):
        print("\n=== Poem Theme Classifier ===")
        print("Enter a poem (type 'exit' to quit):")
        
        while True:
            poem_lines = []
            
            # Get first line
            line = input("> ")
            if line.lower() == 'exit':
                break
            
            # Collect poem lines until empty line
            while line != "":
                poem_lines.append(line)
                line = input("> ")
            
            if poem_lines:
                poem_text = "\n".join(poem_lines)
                result = self.predict(poem_text)
                
                print(f"\nPredicted Theme: {result['theme']}")
                print(f"Confidence: {result['confidence']:.2f}")
                print("\nAll Themes (sorted by probability):")
                
                for theme, prob in result['sorted_probabilities']:
                    print(f"  {theme}: {prob:.4f}")
                
                print("\nEnter another poem (type 'exit' to quit):")

# Create classifier instance
classifier = PoemClassifier(model, label_encoder, tokenizer, max_length)

# Uncomment to run interactive mode
# classifier.interactive_mode()

# Save the classifier state
import pickle
try:
    # Save the label encoder
    with open('label_encoder.pkl', 'wb') as f:
        pickle.dump(label_encoder, f)
    print("Label encoder saved successfully.")
except Exception as e:
    print(f"Error saving label encoder: {e}")
```
